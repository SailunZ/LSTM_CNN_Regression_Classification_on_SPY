```python
import pandas as pd
import numpy as np
import math

import matplotlib.pyplot as plt
from matplotlib.pyplot import figure

import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torch.utils.data import Dataset
from torch.utils.data import DataLoader
from collections import Counter
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler

```

```python
Y_train_raw=pd.read_csv('data/Y_train.csv')
Y_train_raw
```

```python
X_train=np.load('data/X_train_array.npy')  #load X_train array

#construct Y_train_array
#Modify Y_train_raw such that the length matches the length of X_train
Y_train_raw=Y_train_raw.loc[len(Y_train_raw)-X_train.shape[0]:].reset_index(drop=True)
size_y=len(Y_train_raw.index)

#Use the three columns 'hcl','lcl','rcl' classified before according to 'hrate','lrate','rrate'
classes=['<-0.35%','-0.35%~-0.25%','-0.25%~-0.15%','-0.15%~-0.05%','-0.05%~0.05%',
         '0.05%~0.15%','0.15%~0.25%','0.25%~0.35%','>0.35%']

target='hcl'        #set the classification target from 'hcl', 'lcl', or 'rcl' for the model
output_size=9      

if target=='hcl':
    classes=classes[4:]
    output_size=5
elif target=='lcl':
    classes=classes[:5]
    output_size=5

#index to class/class to index dictionary
idx_to_class = {i:j for i, j in enumerate(classes)}
class_to_idx = {value:key for key,value in idx_to_class.items()}

Y_train=np.ndarray((size_y,))
for i in range(size_y):
    Y_train[i]=class_to_idx[Y_train_raw.loc[i][target]]


print(Y_train)

#Pick the latest 10 days for testing. There are 211 data points for each day
n=10
data_x_train = X_train[:-211*n]
data_y_train = Y_train[:-211*n]
data_x_val = X_train[-211*n:]
data_y_val = Y_train[-211*n:]



# normalization
scaler_x = StandardScaler()

data_x_train_s = scaler_x.fit_transform(data_x_train.reshape(-1,data_x_train.shape[-1])).reshape(data_x_train.shape)
data_x_val_s = scaler_x.transform(data_x_val.reshape(-1,data_x_val.shape[-1])).reshape(data_x_val.shape)
print(data_x_val_s[:1]) #first element in data_x_val_s
```

```python
#check Cuda
device = torch.device('cuda:0' if torch.cuda.is_available() else 'cpu')

print(device)
```

```python
#Dataset and Dataloader
class TimeSeriesDataset(Dataset):
    def __init__(self, x, y):
        self.x = x.astype(np.float32)
        self.y = y

    def __len__(self):
        return len(self.x)

    def __getitem__(self, idx):
        return (self.x[idx], self.y[idx])


dataset_train = TimeSeriesDataset(data_x_train_s, data_y_train)
dataset_val = TimeSeriesDataset(data_x_val_s, data_y_val)

print("Train data shape", dataset_train.x.shape, dataset_train.y.shape)
print("Validation data shape", dataset_val.x.shape, dataset_val.y.shape)

batch_size=128

train_dl = DataLoader(dataset_train, batch_size=batch_size, shuffle=True)
val_dl = DataLoader(dataset_val, batch_size=batch_size, shuffle=True)
```

```python
#LSTM model
class LSTMModel(nn.Module):
    def __init__(self, input_size, hidden_size, num_layers, num_classes):
        super().__init__()
        self.lstm = nn.LSTM(input_size=input_size, hidden_size=hidden_size, num_layers=num_layers, batch_first=True)  
        self.linear_1 = nn.Linear(num_layers * hidden_size, 128)
        self.relu = nn.ReLU()      
        self.dropout = nn.Dropout(0.2)
        self.linear_2 = nn.Linear(128, num_classes)
        
    def forward(self, x):
        batchsize = x.shape[0]

        # LSTM layer
        lstm_out, (h_n, c_n) = self.lstm(x)

        # reshape output from hidden cell into [batch, features] for `linear_2`
        x = h_n.permute(1, 0, 2).reshape(batchsize, -1)
        x = self.relu(x)
        x = self.dropout(x)        
        # layer 1
        x = self.linear_1(x)
        x = self.relu(x)
        x = self.dropout(x)
        # layer 2       
        predictions = self.linear_2(x)

        return predictions
```

```python
#Parallel structure of CNN and LSTM
class CNNLSTM(nn.Module):
    def __init__(self, input_size, hidden_size, num_layers, num_classes):
        super(CNNLSTM, self).__init__()
        self.cnn = nn.Sequential(
            nn.Conv1d(in_channels=input_size, out_channels=64, kernel_size=3, stride=1, padding=1),
            nn.BatchNorm1d(64),
            nn.ReLU(),
            nn.MaxPool1d(kernel_size=2, stride=2),
            nn.Conv1d(in_channels=64, out_channels=128, kernel_size=3, stride=1, padding=1),
            nn.BatchNorm1d(128),
            nn.ReLU(),
            nn.MaxPool1d(kernel_size=2, stride=2),
            nn.Flatten(),
            nn.Dropout(0.3),
            nn.Linear(30*128,128),
            nn.BatchNorm1d(128),
            nn.ReLU()
        )
        self.fc1 = nn.Linear(input_size, hidden_size)
        self.lstm = nn.LSTM(input_size=hidden_size, hidden_size=hidden_size, num_layers=num_layers, batch_first=True)
        self.fc_lstm = nn.Linear(hidden_size, 128)
        self.fc2 = nn.Linear(128*2, 64)
        self.fc3 = nn.Linear(64, num_classes)

    def forward(self, x):
        # cnn takes input of shape (batch_size, channels, seq_len)
        x_cnn = x.permute(0, 2, 1)
        out_cnn = self.cnn(x_cnn)
        # lstm takes input of shape (batch_size, seq_len, input_size)
        x_lstm = self.fc1(x)
        out_lstm, _ = self.lstm(x_lstm)
        out_lstm = self.fc_lstm(out_lstm[:, -1, :])
        out_lstm = nn.Dropout(0.3)(out_lstm)
        out = torch.cat([out_cnn, out_lstm], dim=1)
        out = self.fc2(out)
        out = nn.Dropout(0.3)(out)
        out = self.fc3(out)
        return out
```

```python
#CNN first, then LSTM
class CNN_LSTM(nn.Module):
    def __init__(self, input_size, hidden_size, num_layers, num_classes):
        super(CNN_LSTM, self).__init__()
        self.cnn = nn.Sequential(
            nn.Conv1d(in_channels=input_size, out_channels=64, kernel_size=3, stride=1, padding=1),
            nn.BatchNorm1d(64),
            nn.ReLU(),
            nn.MaxPool1d(kernel_size=2, stride=2),
            nn.Conv1d(in_channels=64, out_channels=128, kernel_size=3, stride=1, padding=1),
            nn.BatchNorm1d(128),
            nn.ReLU(),
            nn.MaxPool1d(kernel_size=2, stride=2),
            
        )
        self.lstm = nn.LSTM(input_size=128, hidden_size=hidden_size, num_layers=num_layers, batch_first=True)
        self.fc_lstm = nn.Linear(hidden_size, 128)
        self.fc1 = nn.Linear(128, 64)
        self.fc2 = nn.Linear(64, num_classes)

    def forward(self, x):
        #cnn takes input of shape (batch_size, channels, seq_len)
        x = x.permute(0, 2, 1)
        out = self.cnn(x)        
        # lstm takes input of shape (batch_size, seq_len, input_size)
        out = out.permute(0, 2, 1)
        out_lstm, _ = self.lstm(out)
        out = self.fc_lstm(out_lstm[:, -1, :])
        out = nn.Dropout(0.3)(out)
        out = self.fc1(out)
        out = nn.Dropout(0.3)(out)
        out = self.fc2(out)

        return out
```

```python
#LSTM first, then CNN
class LSTM_CNN(nn.Module):
    def __init__(self, input_size, hidden_size, num_layers, num_classes):
        super(LSTM_CNN, self).__init__()
        self.lstm = nn.LSTM(input_size=hidden_size, hidden_size=hidden_size, num_layers=num_layers, batch_first=True)
        self.cnn = nn.Sequential(
            nn.Conv1d(in_channels=hidden_size, out_channels=64, kernel_size=3, stride=1, padding=1),
            nn.BatchNorm1d(64),
            nn.ReLU(),
            nn.MaxPool1d(kernel_size=2, stride=2),
            nn.Conv1d(in_channels=64, out_channels=128, kernel_size=3, stride=1, padding=1),
            nn.BatchNorm1d(128),
            nn.ReLU(),
            nn.MaxPool1d(kernel_size=2, stride=2),
            nn.Flatten(),
            nn.Dropout(0.3),
            nn.Linear(30*128,128),
            nn.BatchNorm1d(128),
            nn.ReLU()
        )
        self.fc1 = nn.Linear(input_size, hidden_size)
        self.lstm = nn.LSTM(input_size=hidden_size, hidden_size=hidden_size, num_layers=num_layers, batch_first=True)
        self.fc2 = nn.Linear(128, 64)
        self.fc3 = nn.Linear(64, num_classes)

    def forward(self, x):
        x_lstm = self.fc1(x)
        out_lstm, _ = self.lstm(x_lstm)
        out = out_lstm.permute(0, 2, 1)
        out = self.cnn(out)
        out = self.fc2(out)
        out = nn.Dropout(0.3)(out)
        out = self.fc3(out)
        return out
```

```python
# check accuracy for dataloader
def accuracy(dl,classes, model, device): 

    correct = 0
    total = 0

    correct_pred = {classname: 0 for classname in classes}
    total_pred = {classname: 0 for classname in classes}
    pred={classname: 0 for classname in classes}
    
    model.eval()
    
    with torch.no_grad():
        for data in dl:
            images, labels = data[0].to(device), data[1].to(device)
            outputs = model(images)
            _, predictions = torch.max(outputs, 1)
            total += labels.size(0)
            correct += (predictions == labels).sum().item()

            for label, prediction in zip(labels, predictions):
                if label == prediction:
                    correct_pred[classes[int(label)]] += 1
                total_pred[classes[int(label)]] += 1
                pred[classes[int(prediction)]] += 1

    print(f'Accuracy of the network: {100 * correct / total} %\n')
            
    for classname, correct_count in correct_pred.items():
        if total_pred[classname]!=0:
            accuracy = 100 * float(correct_count) / total_pred[classname]
            print(f'Accuracy for class: {classname} is {accuracy:.1f} %')
        else:
            print('No datapoint in the class:',classname)
        print('number of datapoints in the class:',total_pred[classname])
        print('number of datapoints predicted to be in the class:',pred[classname],'\n')
        
    return 100 * correct / total
```

```python
#Trainning
input_size=7 # we are using open, volume, MACD, Signal, RSI, variance, crossover as input features
hidden_size=32
num_layers=2
num_classes=len(classes)

#Choose the model
cnnlstm=CNNLSTM(input_size, hidden_size, num_layers, num_classes).to(device)
#cnn_lstm=CNN_LSTM(input_size, hidden_size, num_layers, num_classes).to(device)
#lstm_cnn=LSTM_CNN(input_size, hidden_size, num_layers, num_classes).to(device)
#lstm=LSTMModel(input_size, hidden_size, num_layers, num_classes).to(device)

model=cnnlstm

#calculate class weights for CrossEntrophyLoss because of imbalanced data
class_count = Counter(Y_train)
class_weights = torch.Tensor([size_y/x[1] for x in sorted(pd.Series(class_count).items(),
                                                                key=lambda x: x[0])]) 
class_weights=class_weights.to(device)

criterion = nn.CrossEntropyLoss(class_weights)

optimizer = torch.optim.Adam(model.parameters(), lr=0.01, betas=(0.9, 0.98), eps=1e-9)
                             

train_rec, val_rec=[], []

for epoch in range(100):  

    running_loss = 0.0
    for i, data in enumerate(train_dl):

        inputs, labels = data[0].to(device), data[1].type(torch.LongTensor).to(device)

        optimizer.zero_grad()

        outputs = model(inputs)
        loss = criterion(outputs.contiguous(), labels.contiguous())
        loss.backward()
        optimizer.step()

        running_loss += loss.item()

        if i%10==0:
            print(f'[{epoch + 1}, {i + 1:4d}] loss: {running_loss:.3f}')
        running_loss = 0.0
        
    train_rec.append(accuracy(train_dl,classes,model,device))
    val_rec.append(accuracy(val_dl,classes,model,device))
    



print('Finished Training')
```

```python
#Prediction on trainning data
plt.plot(range(1,epoch+2), train_rec)
plt.xlabel('epoch')
plt.ylabel('train accuracy(%)')
accuracy(train_dl,classes,model,device)
```

```python
#Prediction on validation data
plt.plot(range(1,epoch+2), val_rec)
plt.xlabel('epoch')
plt.ylabel('test accuracy(%)')
accuracy(val_dl,classes,model,device)
```
